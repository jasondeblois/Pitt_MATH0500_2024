```python
def IsPrime(n):
  for j in range(2,n):
    if n%j == 0:
      return False
  return True
```

```python
IsPrime(51)
```

```python
def goldbach(n):
  for p in range(2,n):
    if IsPrime(p) and IsPrime(n-p):
      return p
  return 'Congratulations, you have a counterexample'
```

```python
max_p = 2
for n in range(4,501,2):
  p = goldbach(n)
  print(n,p)
  if max_p < p:
    max_p = p
print(max_p)
```

```python
def PrimesTo(n):
  L = []
  for j in range(2,n):
    if IsPrime(j):
      L.append(j)
  return L
```
